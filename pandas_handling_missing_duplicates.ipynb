```python
df = pd.read_csv('datasets/titanic.csv')
```

```python
#replace all missing values with 0
df2 = df.fillna(0)
```

```python
df2
```

```python
df2.isna().any()
```

```python
#replacing missing with avg
df3 = df.fillna(df.mean(numeric_only=True))
```

```python
df3
```

```python
#for non numerical columns 
#replacing with most frequent item
df4 = df3.fillna(df.mode().iloc[0])
```

```python
df4
```

```python
df4.isna().any()
```

```python
#check duplicated data
df4.duplicated()
```

```python
df4.duplicated().any()
```
