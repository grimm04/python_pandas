```python

df_pivot = df.pivot_table(index=["id_prov", "id_partai","prov", "partai"], 
                          values="suara_partai",
                          aggfunc="sum").reset_index()
  
```

```python
output_json = df_pivot.groupby('id_prov').apply(lambda x: x.set_index('id_partai')['suara_partai'].to_dict()).to_dict()

print(output_json)
```
